### Utility function

#### Function for cleaning

```python
def manage_title_credits(data, credits):
  credits_group = credits.groupby(["id", "role"])
  actor_character_list, director_list = [], []
  for key, dataframe in credits_group:
    temp_list = []
    temp_list.append(key[0])
    if dataframe["role"].iloc[0] == "DIRECTOR":
      temp_list.append(sorted(dataframe["name"].values.tolist()))
      director_list.append(temp_list)
    else:
      temp_list.append(sorted(dataframe["name"].values.tolist()))
      temp_list.append(sorted(dataframe["character"].values.tolist()))
      actor_character_list.append(temp_list)
  actor_character_dataframe = pd.DataFrame(actor_character_list,
                                           columns=["id", "actor", "character"])
  director_dataframe = pd.DataFrame(director_list, columns=["id", "director"])
  merge_credits = actor_character_dataframe.merge(director_dataframe, on='id')
  merge_credits_data = data.merge(merge_credits, on="id")
  concat_credits_data = pd.concat([data, merge_credits_data], axis=0)
  concat_credits_data.drop_duplicates(subset=["id"], keep="last", inplace=True)
  return concat_credits_data


def rename_title(data):
  column_name_dict = {"production_countries": "production_country",
                      "genres": "genre", "seasons": "season",
                      "imdb_votes": "imdb_vote"}
  change1 = data.rename(columns=column_name_dict)
  name_list = []
  for name in change1["title"]:
    name = name.lower().replace(" - imdb", "")
    name_list.append(name)
  change1["title"] = name_list
  return change1


def fill_title(data):
  fill_na_value_title = {"actor": "[]", "character": "[]", "director": "[]"}
  data = data.fillna(value=fill_na_value_title)
  return data


def float_to_integer_title(data):
  change2 = data.astype({"season": int, "imdb_vote": int})
  return change2


def string_to_list(data):
  def class_list(data):
    data_list = []
    if data != "[]" and type(data) != list:
      tran = data.translate(cleaning_dict)
      split = sorted(tran.split(", "))
      data_list.append(split)
    elif type(data) != list:
      data_list.append([""])
    else:
      data_list.append(data)
    return data_list

  genre_list, country_list, language_list, actor_list = [], [], [], []
  director_list, composer_list, writer_list, character_list = [], [], [], []
  genre = data["genre"]
  country = data["production_country"]
  language = data["language"]
  actor = data["actor"]
  director = data["director"]
  composer = data["composer"]
  writer = data["writer"]
  character = data["character"]
  zipping = zip(genre, country, language, actor, director, composer, writer, character)
  for gen, cty, lan, act, dir, com, wrt, cha in zipping:
    genre_list += class_list(gen)
    country_list += class_list(cty)
    language_list += class_list(lan)
    actor_list += class_list(act)
    director_list += class_list(dir)
    composer_list += class_list(com)
    writer_list += class_list(wrt)
    character_list += class_list(cha)
  data["genre"] = genre_list
  data["production_country"] = country_list
  data["language"] = language_list
  data["actor"] = actor_list
  data["director"] = director_list
  data["composer"] = composer_list
  data["writer"] = writer_list
  data["character"] = character_list
  return data


def replace_title(data):
  genre = data["genre"]
  character = data["character"]
  country = data["production_country"]
  data["genre"] = [[genre_dict.get(g) if g in (genre_dict.keys()) else g for g in gen] for gen in genre]
  data["character"] = [list(filter(lambda c: c != "-", cha)) for cha in character]
  country = [[country_dict.get(c).lower() if c in (country_dict.keys()) else c.lower() for c in cty] for cty in country]
  data["production_country"] = [[country_dict.get(c) if c in (country_dict.keys()) else c for c in cty] for cty in country]
  return data


def arrange(data):
  data = data[["id", "title", "description", "runtime", "release_year",
               "season", "type", "age_certification", "genre", "language",
               "production_country", "character", "actor", "director", "writer",
               "composer", "imdb_score", "imdb_vote"]]
  data = data.reset_index()
  data["id"] = data.index + 1
  data["id"] = data["id"].apply(lambda x: x)
  if data.shape[1] == 19:
    data.drop(["index"], axis=1, inplace=True)
  return data


def manage_case(data):
  data["genre"] = data["genre"].apply(tuple)
  data["production_country"] = data["production_country"].apply(tuple)
  data["language"] = data["language"].apply(tuple)
  data["actor"] = data["actor"].apply(tuple)
  data["director"] = data["director"].apply(tuple)
  data["composer"] = data["composer"].apply(tuple)
  data["writer"] = data["writer"].apply(tuple)
  data["character"] = data["character"].apply(tuple)
  return data
```

### Cleaning imdb dataset (Netflix Prize Shows Information (9000 Shows))

Task List for imdb dataset (complete)
  - Drop na in title
  - Add column
    - description
    - age_certification
    - season
  - Replace "NaN" to "-"
  - Change column name
    - year -> release_year
    - kind -> type
    - rating -> imdb_score
    - vote -> imdb_vote
    - country -> production_country
    - cast -> actor
  - Turn every string to lower
  - Change data type
    - year, imdb_vote, runtime: float -> int
    - genre, production_country, language, actor, director, composer, writer: str -> list
  - Replace type from dictionary
  - Replace language
    - none -> ""
    - old -> "old language"

```python
def drop_imdb(data):
  drop = data.dropna(subset=["Unnamed: 0", 'title'])
  drop = drop[drop["kind"] != "video game"]
  return drop


def add_imdb(data):
  data[["description", "age_certification"]] = "-"
  data["season"] = 0
  data["character"] = "[]"
  return data


def rename_imdb(data):
  column_name_dict = {"Unnamed: 0": "id", "year": "release_year",
                      "kind": "type", "rating": "imdb_score",
                      "vote": "imdb_vote", "country" : "production_country",
                      "cast": "actor"}
  change1 = data.rename(columns=column_name_dict)
  name_list = []
  for name in change1["title"]:
    name = name.lower().replace(" - imdb", "")
    name_list.append(name)
  change1["title"] = name_list
  return change1


def fill_imdb(data):
  fill_na_value = {"release_year": 0, "type": "-", "genre": "[]",
                   "imdb_score": 0.0, "imdb_vote": 0.0, "runtime": 0,
                   "production_country": "[]", "language": "[]", "actor": "[]",
                   "director": "[]", "composer": "[]", "writer": "[]"}
  data = data.fillna(value=fill_na_value)
  return data


def lowercase_imdb(data):
  original_title_name = data["title"].values.tolist()
  data["title"] = data["title"].str.lower()
  data["type"] = data["type"].str.lower()
  data["genre"] = data["genre"].str.lower()
  data["production_country"] = data["production_country"].str.lower()
  data["language"] = data["language"].str.lower()
  data["actor"] = data["actor"].str.lower()
  data["director"] = data["director"].str.lower()
  data["composer"] = data["composer"].str.lower()
  data["writer"] = data["writer"].str.lower()
  return original_title_name, data


def float_to_integer_imdb(data):
  change2 = data.astype({"release_year": int, "imdb_vote": int, "runtime": int})
  return change2


def replace_imdb(data):
  language = data["language"]
  genre = data["genre"]
  country = data["production_country"]
  language_list = [[language_dict.get(l) if l in (language_dict.keys()) else l for l in lan] for lan in language]
  genre_list = [[genre_dict.get(g) if g in (genre_dict.keys()) else g for g in gen] for gen in genre]
  country_list = [[country_dict.get(c) if c in (country_dict.keys()) else c for c in cty] for cty in country]
  data["type"] = data["type"].replace(type_dict)
  data["language"] = language_list
  data["genre"] = genre_list
  data["production_country"] = country_list
  return data


def main_imdb(imdb):
  dropping = drop_imdb(imdb)
  adding = add_imdb(dropping)
  renaming = rename_imdb(adding)
  filling = fill_imdb(renaming)
  title_name, turning = lowercase_imdb(filling)
  changing = float_to_integer_imdb(turning)
  changing2 = string_to_list(changing)
  replacing = replace_imdb(changing2)
  arranging = arrange(replacing)
  return arranging


cleaned_imdb = main_imdb(imdb)
cleaned_imdb.to_pickle("cleaned_imdb.pkl")
```
